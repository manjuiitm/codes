What is Time Complexity?
Time complexity measures how the runtime of an algorithm changes as the size of the input grows. It predicts the performance of the algorithm and helps in choosing the best solution for a problem.

Input Size (n): The size of the input to the algorithm.
Efficiency: An efficient algorithm scales well even with large inputs.
Types of Time Complexities with Python Examples
1. Constant Time – O(1)
The execution time does not depend on the input size.
Example: Accessing an element in a list by its index.
def get_element(arr, index):
    return arr[index]  # Always takes constant time

# Example usage
arr = [1, 2, 3, 4, 5]
print(get_element(arr, 2))  # Output: 3
No matter how large the list is, accessing an element at a specific index takes the same time.

2. Logarithmic Time – O(log n)
The runtime increases logarithmically as the input size grows. Common in divide-and-conquer algorithms.
Example: Binary Search.
def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1  # Element not found

# Example usage
arr = [1, 2, 3, 4, 5, 6, 7, 8, 9]
print(binary_search(arr, 6))  # Output: 5
The input size is halved in each step, making the algorithm logarithmic.

3. Linear Time – O(n)
The runtime grows directly in proportion to the input size.
Example: Finding the maximum element in a list.
def find_max(arr):
    max_element = arr[0]
    for num in arr:
        if num > max_element:
            max_element = num
    return max_element

# Example usage
arr = [10, 20, 5, 8, 25]
print(find_max(arr))  # Output: 25
For a list of size n, the loop runs n times.

4. Quadratic Time – O(n²)
The runtime grows quadratically with the input size. Common in algorithms with nested loops.
Example: Printing all pairs of elements in a list.
def print_pairs(arr):
    for i in range(len(arr)):
        for j in range(len(arr)):
            print(f"({arr[i]}, {arr[j]})")

# Example usage
arr = [1, 2, 3]
print_pairs(arr)
If the input size is 3, the loops run 3×3=9 times.

5. Exponential Time – O(2ⁿ)
The runtime doubles with each additional input element. Common in recursive problems like generating all subsets.
Example: Generating all subsets of a list.
def generate_subsets(arr):
    n = len(arr)
    total = 1 << n  # 2^n subsets
    subsets = []
    for i in range(total):
        subset = []
        for j in range(n):
            if i & (1 << j):  # Check if j-th bit is set in i
                subset.append(arr[j])
        subsets.append(subset)
    return subsets

# Example usage
arr = [1, 2, 3]
print(generate_subsets(arr))
# Output: [[], [1], [2], [1, 2], [3], [1, 3], [2, 3], [1, 2, 3]]
For a list of size n, there are 2^n subsets.

Why Time Complexity Matters?
Scalability: Predict how an algorithm performs with large inputs.
Efficiency: Helps compare and choose the best solution.
Optimization: Guides efforts to improve performance.
Common Time Complexities
Time Complexity	Example Algorithm	Behavior
O(1)	Accessing an element	Constant, unaffected by input size.
O(log n)	Binary search	Fast, input size halved each step.
O(n)	Linear search	Grows directly with input size.
O(n log n)	Merge Sort, Quick Sort	Efficient for sorting.
O(n²)	Bubble Sort, Nested loops	Slow for large inputs.

Quadratic Time (O(n²)): You compare every book with every other book to find duplicates.

What is Recursion?
Recursion is a programming technique where a function calls itself directly or indirectly to solve a problem. It is particularly useful for problems that can be divided into smaller subproblems of the same type. The recursive function typically has two main components:

Base Case: This is the condition under which the recursion terminates. Without a base case, the recursion would run indefinitely.
Recursive Case: This defines how the problem is broken down into smaller instances and solved recursively.
How Does Recursion Work?
When a recursive function is called, the current state of the function (including its local variables and the instruction pointer) is stored in the call stack. Once the base case is reached, the function starts returning values, and the stack unwinds.

Key Characteristics of Recursive Functions:
They must have a base case to prevent infinite recursion.
Each recursive call should move closer to the base case.
They use stack memory for storing function states.
Example: Factorial of a Number
Problem: Calculate the factorial of a non-negative integer n. The factorial of n (denoted as n!) is defined as:

Recursive Function:
#include <stdio.h>

int factorial(int n) {
    if (n == 0 || n == 1) { // Base case
        return 1;
    }
    return n * factorial(n - 1); // Recursive case
}

int main() {
    int num = 5;
    printf("Factorial of %d is %d\n", num, factorial(num));
    return 0;
}


Step-by-Step Explanation:
Suppose we call factorial(5). Here’s how the recursion works:

First Call: factorial(5) Since n != 0, compute 5 * factorial(4) and call factorial(4).
Second Call: factorial(4) Since n != 0, compute 4 * factorial(3) and call factorial(3).
Third Call: factorial(3) Since n != 0, compute 3 * factorial(2) and call factorial(2).
Fourth Call: factorial(2) Since n != 0, compute 2 * factorial(1) and call factorial(1).
Fifth Call: factorial(1) Since n != 0, compute 1 * factorial(0) and call factorial(0).
Base Case: factorial(0) Since n == 0, return 1. This stops the recursion.
Return Process (Unwinding the Recursion):
Now the function starts returning values back to previous calls:

factorial(0) returns 1. So, factorial(1) = 1 * 1 = 1.
factorial(1) returns 1. So, factorial(2) = 2 * 1 = 2.
factorial(2) returns 2. So, factorial(3) = 3 * 2 = 6.
factorial(3) returns 6. So, factorial(4) = 4 * 6 = 24.
factorial(4) returns 24. So, factorial(5) = 5 * 24 = 120.
Final result: factorial(5) = 120.

Visualizing the Call Stack:

    -> factorial(4)
        -> factorial(3)
            -> factorial(2)
                -> factorial(1)
                    -> factorial(0) = 1 (Base case)
                <- factorial(1) = 1 * 1 = 1
            <- factorial(2) = 2 * 1 = 2
        <- factorial(3) = 3 * 2 = 6
    <- factorial(4) = 4 * 6 = 24
<- factorial(5) = 5 * 24 = 120
This step-by-step breakdown shows how recursion simplifies solving the factorial problem by delegating work to smaller instances of itself.

Dynamic Programming
Dynamic Programming (DP) is an optimization technique used to solve problems with overlapping subproblems and optimal substructure. It is widely used in solving complex problems efficiently by breaking them into simpler subproblems.

Key Concepts of Dynamic Programming
1. Overlapping Subproblems
Definition: A problem exhibits overlapping subproblems if it can be broken down into smaller subproblems that are solved multiple times during the computation.

Key Characteristics:
The same subproblem appears repeatedly.
To optimize, the results of these subproblems can be stored (memoization or tabulation) to avoid redundant calculations.
Example: Fibonacci Numbers
The Fibonacci sequence is defined as:


If we compute F(5) recursively, the tree of function calls looks like this:

           F(5)
          /    \
       F(4)    F(3)
      /   \    /   \
   F(3)  F(2) F(2) F(1)
  /   \  /   \
F(2) F(1) F(1) F(0)
Observation:

F(3) is computed twice.
F(2) is computed three times.
The same subproblems (F(3), F(2)) are solved repeatedly, leading to redundant computations.
Solution: Store the results of F(3) and F(2) after computing them once. This eliminates the need to recompute these values.

Fibonacci with Memoization
def fibonacci(n, memo={}):
    if n in memo:
        return memo[n]
    if n <= 1:
        return n
    memo[n] = fibonacci(n - 1, memo) + fibonacci(n - 2, memo)
    return memo[n]

# Example usage
print(fibonacci(5))  # Output: 5


2. Optimal Substructure
Definition: A problem exhibits optimal substructure if the solution to the problem can be constructed from the optimal solutions of its subproblems.

Key Characteristics:
The global optimal solution can be derived from the optimal solutions of smaller subproblems.
This property is essential for applying dynamic programming.
Example: Shortest Path in a Graph
Consider a graph where you want to find the shortest path from node A to node D.

Graph:

A --1--> B
A --4--> C
B --2--> D
C --1--> D
Problem Statement: Find the shortest path from A to D.

Step 1: Break it into subproblems:

The shortest path from A to D

can either:

Go through B (A → B → D).
Go through C (A → C → D).
Step 2: Solve subproblems:

Shortest path from A → B = 1, and from B → D = 2. So, A → B → D = 1 + 2 = 3.
Shortest path from A → C = 4, and from C → D = 1. So, A → C → D = 4 + 1 = 5.
Step 3: Combine subproblem solutions:

The shortest path from A to D is the minimum of the two paths:


Optimal Substructure Property:
The shortest path from A to D was constructed from:

The shortest path from A to intermediate nodes (B or C).
The shortest path from intermediate nodes (B or C) to D.




Steps to Solve a Problem Using Dynamic Programming
Define the Problem and Identify Subproblems:

Break the problem into smaller overlapping subproblems.
Decide the State:

The state is the smallest amount of information required to describe a subproblem.
Example: In the Fibonacci problem, the state can be defined as dp[i], the ith Fibonacci number.
Write the Recurrence Relation:

Express the solution of the current problem in terms of its subproblems.
Example: For Fibonacci: dp[i] = dp[i-1] + dp[i-2].
Choose Top-Down (Recursion + Memoization) or Bottom-Up (Tabulation):

Top-Down: Solve the problem recursively while storing results (memoization).
Bottom-Up: Solve all subproblems iteratively in a specific order and store their results (tabulation).
Implement Base Cases:

Initialize the smallest subproblem(s) directly.
Top-Down (Memoization) vs. Bottom-Up (Tabulation)
Feature	Top-Down (Memoization)	Bottom-Up (Tabulation)
Approach	Recursive	Iterative
Storage	Stores results in a dictionary or array	Precomputes results in an array
Ease of Coding	Easy to implement directly from recurrence	Requires planning the computation
Time Complexity	Same as Bottom-Up (depends on the recurrence)	Same as Top-Down
Space Complexity	Higher due to recursion stack	Lower (no recursion stack)
Dynamic Programming Techniques
1. Memoization (Top-Down Approach)
Store the results of previously computed subproblems in a table (array or dictionary).
Avoid redundant calculations by checking if a solution exists before recomputing.
Example: Fibonacci Numbers (Memoization)

2. Tabulation (Bottom-Up Approach)
Solve the problem iteratively and store results in a table.
Fill the table starting from base cases, proceeding towards the desired result.
Example: Fibonacci Numbers (Tabulation)
def fibonacci(n):
    dp = [0] * (n + 1)
    dp[1] = 1
    for i in range(2, n + 1):

    return dp[n]

Problem Statement: Frog Jump with Minimum Cost
A frog is trying to cross a line of stones. Each stone is at a certain height, represented by the array heights. The frog initially at the first stone and aims to reach the last stone with the minimum total cost.

The cost to jump from one stone to another is the absolute difference in their heights. The frog can jump to:

1. The next stone, or
2. The next-to-next stone.
Write a function minCost(heights) that returns the minimum total cost for the frog to reach the last stone.

Example:
Input:

heights = [10, 30, 40, 20]
Output:

30
Explanation:

The frog jumps from the first stone (height 10) to the second stone (height 30), and then to the fourth stone (height 20).

Jump costs:

From 10 to 30: |30 - 10| = 20
From 30 to 20: |30 - 20| = 10
Total cost: 20 + 10 = 30.

```python
# Write Solution Function Here
def minCost(heights):
    n = len(heights)
    dp = [0] * n
    dp[0] = 0

    for i in range(1, n):
        one_step = dp[i - 1] + abs(heights[i] - heights[i - 1])
        two_step = dp[i - 2] + abs(heights[i] - heights[i - 2]) if i > 1 else float('inf')
        dp[i] = min(one_step, two_step)

    return dp[-1]

heights = eval(input())
print(minCost(heights))
```

Problem Statement: Climbing Stairs

You are climbing a staircase that has
 stairs, and you can either take 1 or 2 steps at a time to climb on stairs. Write a function climb_stairs(n) to determine how many distinct ways there are to reach the top of the staircase.
Requirements:

The input
 represents the number of stairs.
Each step can either be taken as a single step (1 step) or as a double step (2 steps).
Use Dynamic Programming to solve the problem efficiently.
Constraints:





Example:

Input 1: n = 2

Output: 2

Explanation:

Take two 1-steps: (1, 1)
Take one 2-step: (2) Hence, there are 2 distinct ways to climb 2 steps.


Input 2: n = 3

Output: 3

Explanation:

Take three 1-steps: (1, 1, 1)
Take one 2-step followed by a 1-step: (2, 1)
Take one 1-step followed by a 2-step: (1, 2) Hence, there are 3 distinct ways to climb 3 steps.

```python
def climb_stairs(n):
    if n <= 1:
        return 1
    st = [0] * (n + 1)
    st[0], st[1] = 1, 1
    for i in range(2, n + 1):
        st[i] = st[i - 1] + st[i - 2]
    return st[n]

n = int(input())
print(climb_stairs(n))
```

Problem Statement: Maximum Sum of Non-Adjacent Numbers

Given an array of integers arr, find the maximum sum of elements such that no two selected elements are adjacent.

Write a function maxSum(arr) that returns the maximum sum of elements such that no two selected elements are adjacent.

Example:

Input:

arr = [3, 2, 7, 10]
Output:

13
Explanation:

Selecting 3 and 10 gives a sum of 3 + 10 = 13.
Other combinations like 2 + 7 = 9 result in a smaller sum.
Adjacent elements cannot be selected.

```python
# Write solution function here
def maxSum(arr):
    n = len(arr)
    if n == 0:
        return 0
    if n == 1:
        return arr[0]

    dp = [0] * n
    dp[0] = arr[0]
    dp[1] = max(arr[0], arr[1])

    for i in range(2, n):
        dp[i] = max(dp[i - 1], arr[i] + dp[i - 2])

    return dp[-1]


arr = eval(input())
print(maxSum(arr))
```

Problem Statement:Minimize Steps to Reduce to 1

    2. Divide n by 2 if n is divisible by 2.
    3. Divide n by 3 if n is divisible by 3.

The task is to find the minimum number of operations required to reduce n to 1.

Write a function minOperations(n) that returns the minimum number of operation to reduce n to 1.

Example:
Input:
n = 10
Output:
3
Explanation:
Start with 10.
Subtract 1 → 9.
Divide by 3 → 3.
Divide by 3 → 1.
Total operations: 3.

```python
def minOperations(n):
    ops = [0] * (n + 1)
    ops[1] = 0

    for i in range(2, n + 1):
        ops[i] = ops[i - 1] + 1
        if i % 2 == 0:
            ops[i] = min(ops[i], ops[i // 2] + 1)
        if i % 3 == 0:
            ops[i] = min(ops[i], ops[i // 3] + 1)

    return ops[n]


n = int(input())
print(minOperations(n))
```

Problem Statement: Coin Change Problem

You are given a set of coins in array coins of distinct denominations and an integer amount. Your goal is to find the minimum number of coins needed to make up the given amount. If it is not possible to make the amount using the given coins, return -1.

Write a function minCoins(coins, amount), that returns the minimum number of coins needed to make up the given amount.

Example:

Input:

coins = [1, 2, 5]
amount = 11
Output:

3
Explanation:

The minimum number of coins to make 11 is 3: 5 + 5 + 1 = 11.

```python
# Write Solution Function Here
def minCoins(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0

    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1



coins = eval(input())
amount = int(input())
print(minCoins(coins,amount))
```

Problem Statement: Longest Palindromic Subsequence

Given a string s, find the length of the longest subsequence of s that is a palindrome.

Write a function longestPalindromicSubsequence(s) that returns the length of the longest palindromic subsequence.

Note: A subsequence is a sequence derived from another sequence by deleting some or no elements without changing the order of the remaining elements.

Example:

Input:

s = "bbabcbcab"
Output:

7
Explanation: The longest palindromic subsequence is "babcbab" or "bacbcab", both of length 7.

```python
# Write Solution function here
def longestPalindromicSubsequence(s: str) -> int:
    n = len(s)
    dp = [[0] * n for _ in range(n)]

    for i in range(n):
        dp[i][i] = 1

    for length in range(2, n + 1):  # length of the substring
        for i in range(n - length + 1):
            j = i + length - 1
            if s[i] == s[j]:
                if length == 2:
                    dp[i][j] = 2
                else:
                    dp[i][j] = 2 + dp[i + 1][j - 1]
            else:
                dp[i][j] = max(dp[i + 1][j], dp[i][j - 1])

    return dp[0][n - 1]


s = input()
print(longestPalindromicSubsequence(s))
```

Problem Statement: Resource Allocation in Projects

You are given n projects, where each project i requires a cost cost[i] and yields a profit profit[i]. You have a maximum budget B. Determine the maximum profit you can achieve without exceeding the budget.

Write a function maxProfit(cost, profit, B) that returns the maximum profit you can achieve without exceeding the budget.

Example:

Input:

cost = [1, 3, 4, 5]
profit = [10, 40, 50, 70]
B = 8
Output:

110
Explanation:

By selecting projects with costs 3 and 5 (profits 40 and 70), the total profit is 110, which is the maximum achievable without exceeding the budget of 8.

```python
# Write Solution function here
def maxProfit(cost, profit, B):
    n = len(cost)
    dp = [[0] * (B + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        for b in range(B + 1):
            if cost[i - 1] <= b:
                dp[i][b] = max(
                    dp[i - 1][b],
                    dp[i - 1][b - cost[i - 1]] + profit[i - 1]
                )
            else:
                dp[i][b] = dp[i - 1][b]

    return dp[n][B]

#
cost = eval(input())
profit = eval(input())
B = int(input())
print(maxProfit(cost, profit, B))
```

Problem Statement: Number of unique paths in a grid with obstacles

You have given a grid[m][n]. Assume that we are starting from (0,0) and the goal is to reach (m-1,n-1). Each move is restricted to a 1-step right or 1-step bottom only. The obstacle is marked as 1 and the space is marked as 0 in the grid You need to find the number of unique paths to reach the goal (m-1,n-1).

Write a function uniquePathCount(grid) that return the number of unique paths to reach the goal (m-1,n-1) from (0,0).

Example:

Input:

[[0,0,0],
 [0,1,0],
 [0,0,0]]
Output:

2
Explanation:

There are two ways to reach the bottom-right corner

Right -> Right -> Down -> Down: Start from (0,0) -> (0,1) -> (0,2) -> (1,2) -> (2,2)
Down -> Down -> Right -> Right: Start from (0,0) -> (1,0) -> (2,0) -> (2,1) -> (2,2)


```python
# Write Solution function here
def uniquePathCount(grid):
    m = len(grid)
    n = len(grid[0])

    dp = [[0] * n for _ in range(m)]

    if grid[0][0] == 0:
        dp[0][0] = 1

    for i in range(1, m):
        if grid[i][0] == 0:
            dp[i][0] = dp[i - 1][0]

    for j in range(1, n):
        if grid[0][j] == 0:
            dp[0][j] = dp[0][j - 1]

    for i in range(1, m):
        for j in range(1, n):
            if grid[i][j] == 0:
                dp[i][j] = dp[i - 1][j] + dp[i][j - 1]

    return dp[m - 1][n - 1]

grid = eval(input())
print(uniquePathCount(grid))
```

Longest Decreasing Susequence(LDS)

Find the length of Longest Decreasing Subsequence(LDS) of given list, where the subsequence elements are highest to lowest.

Note:- All elements are distinct.

Write a function longest_decreasing_sequence(L) that accepts list of L values and returns the length of longest decreasing subsequence.

Example

L = [20, 15, 8, 12, 2]

Here longest decreasing subsequence is [20, 15, 8, 2] and it's length is 4.

Input 1

[20,15,8,12,2]
Output

4
Input 2

[5,10,3]
Output

2



```python
# Write solution function here
from functools import lru_cache

def longest_decreasing_sequence(L):
    n = len(L)

    @lru_cache(maxsize=None)
    def lds_ending_at(i):
        max_len = 1  # At least the element itself
        for j in range(i):
            if L[j] > L[i]:
                max_len = max(max_len, lds_ending_at(j) + 1)
        return max_len

    return max(lds_ending_at(i) for i in range(n)) if n > 0 else 0


arr = eval(input())
print(longest_decreasing_sequence(arr))
```

Fourth Move: Index 3 to index 4

```python
# Write solution function here
def min_moves_to_end(steps):
    n = len(steps)
    if n <= 1:
        return 0

    moves = 0
    current_end = 0
    farthest = 0

    for i in range(n - 1):  # No need to jump from last index
        farthest = max(farthest, i + steps[i])
        if i == current_end:
            moves += 1
            current_end = farthest
            if current_end >= n - 1:
                break

    return moves


steps = eval(input())
print(min_moves_to_end(steps))
```

Minimize cost to reach the destination

You are given a grid of size m x n with positive integer values representing the cost you have to pay in each cell. You can only jump either one cell down or one cell right at a point in time. Your goal is to find the minimum cost to reach the bottom-right corner cell (m-1, n-1) of the grid, starting from the top-left corner cell (0, 0).

Write a function min_cost(m, n, grid_costs) that takes in two integers m and n representing the number of rows (indexed from 0 to m-1) and the number of columns (indexed from 0 to n-1), and a 2D nested list grid_costs representing the cost of each cell.

The function should return an integer indicating the minimum cost to reach the bottom-right corner cell (m-1, n-1) of the grid starting from the top-left corner cell (0, 0).

Sample Input 1

3 #n
[[1, 3, 1],[1, 5, 1],[4, 2, 1]] #grid_costs
Output

7
Explanation

[[1, 3, 1],
 [1, 5, 1],
 [4, 2, 1]]
The minimum cost path is 1(0,0) -> 3(0,1) -> 1(0,2) -> 1(1,2) -> 1(2,2), with a total cost of 7.

Output

Explanation

[[1, 3, 2],
 [1, 2, 4],
 [4, 1, 1]]
The minimum cost path is 1(0,0) -> 1(1,0) -> 2(1,1) -> 1(2,1) -> 1(2,2), with a total cos

```python
# Write solution function here
def min_cost(m, n, grid_costs):
    # Create a DP table initialized with 0s
    dp = [[0 for _ in range(n)] for _ in range(m)]

    # Initialize the starting cell
    dp[0][0] = grid_costs[0][0]

    # Fill the first row (can only come from the left)
    for j in range(1, n):
        dp[0][j] = dp[0][j - 1] + grid_costs[0][j]

    # Fill the first column (can only come from above)
    for i in range(1, m):
        dp[i][0] = dp[i - 1][0] + grid_costs[i][0]

    # Fill the rest of the table
    for i in range(1, m):
        for j in range(1, n):
            dp[i][j] = min(dp[i - 1][j], dp[i][j - 1]) + grid_costs[i][j]

    return dp[m - 1][n - 1]


m = int(input())
n = int(input())
grid_costs = eval(input())
print(min_cost(m, n, grid_costs))
```
